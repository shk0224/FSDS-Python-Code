```python
#Login Function demo- dswithshuvo.com

""""
Actual Credentials:

Email: Shuvo.du333@gmail.com
Password: Shuvo@12345
"""

email= input("Enter your email:       ")
password=input("Enter your password:        ")

actual_email="Shuvo.du333@gmail.com"
actual_password="Shuvo@12345"

if email==actual_email and password==actual_password:
    print('Login Successful')

elif email==actual_email and password!=actual_password:
    print("Wrong Password!")

else:
    print("Invalid Credentials!")
```

# Nested Conditional Statement

```python
#Login Function demo- dswithshuvo.com

""""
Actual Credentials:

Email: Shuvo.du333@gmail.com
Password: Shuvo@12345
"""

email= input("Enter your email:       ")
password=input("Enter your password:        ")

actual_email="Shuvo.du333@gmail.com"
actual_password="Shuvo@12345"

if email==actual_email and password==actual_password:
    print('Login Successful')

elif email==actual_email and password!=actual_password:
    print("Please try again!")
    password=input("Enter your password:        ")

    if password==actual_password:
        print("Success!")
    else:
        print("Wrong password again!")


else:
    print("Invalid Credentials!")
```
